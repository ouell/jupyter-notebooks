```
def login_hypera():
  #Faz o login e aguarda a próxima tela
  driver.get("https://hypera.salesfarma.com.br/PortalConfig/Account/Login")
  WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, '//*[@id="mws-login-form"]/form/div[5]/input')))
  driver.execute_script("document.getElementById('UserName').value = '';")
  driver.execute_script("document.getElementById('Password').value = '';")
  driver.find_element_by_xpath('//*[@id="mws-login-form"]/form/div[5]/input').click()
  WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, '//*[@id="btnPesquisar"]')))
```

```
def pesquisa_inicial_hypera():
  driver.find_element_by_xpath('//*[@id="chkRetornoAntesDe"]').click()
  driver.find_element_by_xpath('//*[@id="btnPesquisar"]').click()
  WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, '//*[@id="pendencias_filter"]/label/input')))
```
